```python
from scipy.ndimage.filters import gaussian_filter
import numpy as np
import pylab as plt
import mrcfile
```

```python
k= mrcfile.open(r'C:\Users\mim4677\Documents\MCB 111\data\Falcon_2012_06_12-14_33_35_0.mrc')
k.data
```

```python
f = plt.figure()
f.add_subplot(1,2, 1)
plt.imshow(k.data[100:3996,100:3996])
#close-up
f.add_subplot(1,2, 2)
plt.imshow(k.data[100:1000,100:1000])
```

```python
#implementing Gaussian image filter; trial and error with 
blurred = gaussian_filter(k.data, sigma = 20)
```

```python
g = plt.figure()
g.add_subplot(1,2, 1)
plt.imshow(blurred[100:3996,100:3996])
#close-up
g.add_subplot(1,2, 2)
plt.imshow(blurred[100:1000,100:1000])
```

```python
for i in range(100,3000):
    for j in range(100,3000):
        if(blurred[i][j] < 70500):
            blurred[i][j]=68000
        else:
            blurred[i][j] = 70500
plt.imshow(blurred[100:3000,100:3000])   
```
